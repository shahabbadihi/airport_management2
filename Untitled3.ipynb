```python
import hashlib
import json
from IPython.display import clear_output

clear_output()

file_Genesis_Block = "/content/blockchain/GenesisBlock.json"


with open(file_Genesis_Block, "rb") as f:
  bytes_ = f.read()
  hash_Genesis_Block = hashlib.sha256(bytes_).hexdigest()


print("Genesis Block Hash:  ", hash_Genesis_Block, end="\n\n")


dir_ledgers = "/content/blockchain/Ledgers/"
dir_math_problems = "/content/blockchain/Math_Problems/"

prev_hash = hash_Genesis_Block


for i in range(2, 17):

  file_ledger = dir_ledgers + "Ledger_Number" + str(i) + ".json"
  file_math_problem = dir_math_problems + "Math_Problem_Number" + str(i) + ".json"
  # print(file_ledger)
  with open(file_ledger, "rb") as f:
    bytes_ = f.read()
    with open(file_ledger, "r") as ff:
      data_ledger = json.load(ff)
    with open(file_math_problem) as f2:
      nonce_Genesis_Block = 0
      data = json.load(f2)

  while True:
    hash = hashlib.sha256(b"".join([nonce_Genesis_Block.to_bytes(4, 'big'),
                prev_hash.encode(), bytes_])).hexdigest()
    if (hash.endswith(data["mathProblem"])):
      break
    nonce_Genesis_Block += 1

  print("Block No.: ", data_ledger["blockNumber"])
  print("Block Hash:  ", hash)
  print("Prev Hash: ", prev_hash)
  print("Nonce Value: ", nonce_Genesis_Block, end="\n\n\n")

  prev_hash = hash
```
